```python
from transformers import BertTokenizerFast, BertForTokenClassification

tokenizer = BertTokenizerFast.from_pretrained("MrPotato/ner-bert-uncased-geocite")
model = BertForTokenClassification.from_pretrained("MrPotato/ner-bert-uncased-geocite")
```

```python
from colorama import Fore, Back, Style

style_map = {
    'B-PER': f'{Back.RED}{Fore.WHITE}',
    'I-PER': f'{Back.LIGHTRED_EX}{Fore.WHITE}',
    'B-ORG': f'{Back.BLUE}{Fore.WHITE}',
    'I-ORG': f'{Back.LIGHTBLUE_EX}{Fore.WHITE}',
    'B-LOC': f'{Back.GREEN}{Fore.BLACK}',
    'I-LOC': f'{Back.LIGHTGREEN_EX}{Fore.BLACK}',
    'O': f'{Style.RESET_ALL}'
}
```

```python
def print_entities(entities, sentence):
    offset = 0
    for entity in entities:
        sentence = sentence[:entity["start"] + offset] + style_map[entity["entity"]] + sentence[entity["start"] + offset:]
        offset += len(style_map[entity["entity"]])
        sentence = sentence[:entity["end"] + offset] + style_map["O"] + sentence[entity["end"] + offset:]
        offset += len(style_map["O"])
    print(sentence)
```

```python
from transformers import pipeline

classifier = pipeline("ner", model=model, tokenizer=tokenizer)
```

```python
example1 = """
Feine, J.S./Lavigne, G.J./Thuan Dao, T.T./Morin, C./Lund, J.P., 1998: Memories of
chronic pain and perceptions of relief. Pain 77: 137-141.
Festinger, L., 1957: A theory of cognitive dissonance. Stanford: Stanford University
Press.
Goethals, G.R./Reckman, R.F., 1973: The perception of consistency in attitudes.
Journal of Experimental Social Psychology 9: 491-501.
Grau, I./Bierhoff, H.W., 1996: Erinnerungsverzerrung durch implizite Theorien. Der
Zusammenhang zwischen aktuellen Urteilen und Erinnerungen an frühere Urteile
über sich selbst und den Lebenspartner. Zeitschrift für Psychologie 204: 167-185.
Linton, S.J./Melin, L., 1982: The accuracy of remembering chronic pain. Pain 13:
281-285.
McFarland, C./Ross, M., 1987: The relation between current impressions and memories
of self and dating partners. Personality and Social Psychology Bulletin 13: 228-238.
Moller-Jensen, O./Wahrendorf, J./Rosenqvist, A./Geser, A., 1984: The reliability of
questionaire-derived historical dietary information and temporal stability of food
habits in individuals. American Journal of Epidemiology 120: 281-290.
"""
```

```python
entities = classifier(example1)
print_entities(entities, example1)
```

```python
example2 = """
DIE VERZERRUNG VON ERINNERUNGEN DURCH
DAS VORWISSEN DER BEFRAGTEN:
DIE ROLLE IMPLIZITER THEORIEN
INA GRAU, ULRICH MUELLER UND ANDREAS ZIEGLER
Implizite Theorien sind schematische Wissenstrukturen über die Stabilität oder die
Veränderbarkeit von Merkmalen
"""
```

```python
entities = classifier(example2)
print_entities(entities, example2)
```

```python
example3 = """
Extracting literature references in German Speaking
Geography – the GEOcite project
Bastian Birkeneder1, Philipp Aufenvenne1, Christian Haase1, Philipp Mayr2 and
Malte Steinbrink1
1Chair of Human Geography, University of Passau, Germany
2GESIS – Leibniz Institute for the Social Sciences, Cologne, Germany
"""
```

```python
entities = classifier(example3)
print_entities(entities, example3)
```
